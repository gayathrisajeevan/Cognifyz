```python
import pandas as pd

file_path = 'Dataset .csv'
df = pd.read_csv(file_path)
df
```

TASK 2: Votes Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify the restaurants with the highest and lowest number of votes
most_votes = df.loc[df['Votes'].idxmax()]
least_votes = df.loc[df['Votes'].idxmin()]

print("Restaurant with the Highest Number of Votes:")
print(most_votes[['Restaurant Name', 'Votes', 'Aggregate rating']])

print("\nRestaurant with the Lowest Number of Votes:")
print(least_votes[['Restaurant Name', 'Votes', 'Aggregate rating']])

# Analyze the correlation between the number of votes and the rating
correlation = df['Votes'].corr(df['Aggregate rating'])
print(f"\nCorrelation between Votes and Rating: {correlation:.2f}")

# Visualize the correlation using a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=df, alpha=0.6)
plt.xlabel('Number of Votes')
plt.ylabel('Rating')
plt.title('Correlation Between Votes and Rating of Restaurants')
plt.grid(True)
plt.tight_layout()
plt.show()
```

TASK 3: Price Range vs. Online Delivery and Table Booking

```python
# Check unique values in these columns to understand the data
print(df['Price range'].unique())
print(df['Has Online delivery'].unique())
print(df['Has Table booking'].unique())

# Create a summary of services by price range
service_analysis = df.groupby('Price range').agg(
    Online_Delivery_Available=('Has Online delivery', lambda x: (x == 'Yes').mean() * 100),
    Table_Booking_Available=('Has Table booking', lambda x: (x == 'Yes').mean() * 100),
    Restaurant_Count=('Restaurant Name', 'count')
).sort_index()

print("Service Availability by Price Range:")
print(service_analysis)

# Visualization of service availability by price range
plt.figure(figsize=(10, 6))

# Plot for online delivery availability
plt.subplot(1, 2, 1)
sns.barplot(x=service_analysis.index, y='Online_Delivery_Available', data=service_analysis, palette='Blues')
plt.xlabel('Price Range')
plt.ylabel('Percentage Offering Online Delivery')
plt.title('Online Delivery Availability by Price Range')

# Plot for table booking availability
plt.subplot(1, 2, 2)
sns.barplot(x=service_analysis.index, y='Table_Booking_Available', data=service_analysis, palette='Greens')
plt.xlabel('Price Range')
plt.ylabel('Percentage Offering Table Booking')
plt.title('Table Booking Availability by Price Range')

plt.tight_layout()
plt.show()
```
